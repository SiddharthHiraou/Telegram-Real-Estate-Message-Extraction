```python
import pandas as pd
```

```python
file=open('messages.html','r')
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(file,'html.parser')
```

```python
lists=soup.find_all('div',attrs={'class':'text'})
```

```python
lists
```

```python
info_clean=[]
for i in range(len(lists)):
    clean_text=lists[i].get_text()
    info_clean.append(clean_text)
```

```python
info_clean
```

```python
cleaned_data=[]
for line in info_clean:
    linesplit=line.split("!!")
    if len(linesplit)==3:
        location=linesplit[0]
        details=linesplit[1]
        links=linesplit[2]
        cleaned_data.append([location, details,links])
    elif len(linesplit)==2:
        location=linesplit[0]
        details=linesplit[1]
        links=None
        cleaned_data.append([location, details,links])
    
    
```

```python
cleaned_data
```

```python
df = pd.DataFrame(cleaned_data, columns = ['LOCATION','DETAILS','LINKS'])
```

```python
df.to_excel('Tele_Post.xlsx', index=False)
```
